```python
"""
Fungsi ini digunakan untuk pembersihan setiap komentar dalam tweet yang dimana berisi tahapan proses membersihkan simbol-simbol
atau karakter yang berulang dan kurang penting yang tidak diperlukan dalam dataset yang berakibat mengganggu proses klasifikasi 
dengan cara menghilangkan noise/gangguan (simbol bawaan atau tanda baca default dari twitter) agar kualitas data menjadi 
lebih baik sehingga hasil penelitian lebih akurat
"""
def clean_text(reviews):
    all_reviews = list()
    lines = reviews["text"].values.tolist()
    for text in lines:
        text = re.sub("RT "," ", text) #hapus RT 
        text = re.sub("@\\S+", " ", text) #hapus @ 
        text = re.sub(r"http\S+", " ", text) #hapus URL atau incomplete URL
        text = re.sub("#\S+", " ", text) #hapus #
        text = re.sub(r"\d+", " ", text) #hapus angka 1 digit atau lebih
        text = re.sub(r"[-!\"#$%&'()*+,./:;<=>?@\[\\\]_`{|}~ÃƒÆ’Ã‚Â¿]", " ", text) #hapus simbol sama seperti punctuation character
        text = re.sub(r"\n", " ", text) #hapus newline
        text = re.sub(" +", " ",text) #gabung spasi dengan huruf belakangnya
        text = text.translate(str.maketrans(" "," ",string.punctuation)) 
        #remove punctuation atau tanda baca, string berisi huruf, spasi, angka.
        #Dan ada tanda baca: karakter ini termasuk koma dan titik serta titik koma.
        
        text = text.strip() #hapus spasi di depan dan di belakang
        text = re.sub("\s+"," ", text) #hapus beberapa spasi menjadi satu spasi
        text = re.sub(r"\b[a-zA-Z]\b", " ", text) #hapus single character
        all_reviews.append(text) #Setelah semua komentar selesai dibersihkan, komentar atau text di dalam variabel "line" 
        #tersebut akan ditambahkan atau ditaruh ke dalam variabel all_reviews dimana data tersebut akan berupa dalam bentuk list
    return all_reviews
all_reviews = clean_text(data)
print('Cleansing Data Result : \n') 
all_reviews[0:20]


#referensi : https://docs.python.org/3/library/re.html
#referensi : https://www3.ntu.edu.sg/home/ehchua/programming/howto/Regexe.html
#referensi : https://www.dotnetperls.com/punctuation-python
```

```python
#library yang diperlukan
#!pip install Sastrawi
#!pip install nltk
#nltk.download("popular")

#Import library yang dibutuhkan dalam proses pre-processing
import string #library untuk bekerja dengan text
import re #regex library
#import itertools #Library iterator untuk perulangan yang efisien
import nltk #library yang merupakan bagian dari NLP yang untuk bekerja dengan permodelan teks
from nltk.tokenize import word_tokenize #import library package untuk proses Tokenisasi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory #import library package untuk proses Stemming
from nltk.corpus import stopwords #import library package untuk proses Stopwords
#from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
```

```python
# ------ Case Folding --------
#Digunakan untuk transform huruf menjadi huruf lower atau kecil agar seragam
data['sentiment_text'] = data['text'].str.lower()

print('Case Folding Result : \n')
print(data['sentiment_text'].head(20))
print('\n\n\n')

#Fungsi yang digunakan untuk mengganti atau menghapus karakter yang spesial
def remove_text_special(sentiment_text):
    #hapus tab, new line, and back slice
    sentiment_text = sentiment_text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"").replace('\x80Â¦',"")
    #hapus non ASCII (emoticon, chinese word, dll)
    sentiment_text = sentiment_text.encode('ascii', 'replace').decode('ascii')
    return sentiment_text     

data['sentiment_text'] = data['sentiment_text'].apply(remove_text_special)

#hapus emoji âœ ðŸŒ· ðŸ“Œ ðŸ‘ˆðŸ» ðŸ–¥
def remove_emoji(sentiment_text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002500-\U00002BEF"  # chinese char
                           u"\U00002702-\U000027B0"
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u"\U00010000-\U0010ffff"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u200d"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\ufe0f"  # dingbats
                           u"\u3030"
                               #mengonversi data input ke Unicode
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', sentiment_text)

data['sentiment_text'] = data['sentiment_text'].apply(remove_emoji)

# ------ Tokenizing ---------
"""
Fungsi ini digunakan untuk proses tokenisasi yang dimana proses memecah kalimat menjadi kata-kata menjadi lebih berarti 
dan bermakna. Tokenisasi memisah setiap kata yang menyusun kalimat.
"""
def word_tokenize_wrapper(sentiment_text):
    tokens = word_tokenize(sentiment_text)
    
    #proses normalization menghilangkan slangword dari dict appos yang sebelumnya sudah dibuat
    reformed = [appos[word] if word in appos else word for word in tokens]
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in reformed]
    words = [word for word in stripped if word.isalpha()]
    return words

data['sentiment_text'] = data['sentiment_text'].apply(word_tokenize_wrapper)


print('Tokenizing Result : \n') 
print(data['sentiment_text'].head(20))
print('\n\n\n')

```

```python
def negative_handling(sentiment_text):
    negative_sentiment_text = []
    for i in range(len(sentiment_text)):
        word = sentiment_text[i]
        if sentiment_text[i-1] not in ['tidak','kurang','tak','tanpa','terlalu','bukan','belum']:
            negative_sentiment_text.append(word)
        else:
            word = "%s_%s" %(sentiment_text[i-1],word)
            negative_sentiment_text.append(word)
    return negative_sentiment_text

data['sentiment_text'] = data['sentiment_text'].apply(negative_handling)
print('Negation Result : \n') 
print(data['sentiment_text'].head(20))
print('\n\n\n')
```

```python
"""
Pada bagian ini merupakan proses untuk mengubah atau mengembalikan kata pada setiap kalimat ke bentuk dasar atau menghapus 
kata-kata imbuhan.Proses menemukan kata dasar (root word) dari kata berimbuhan (affixed word) dengan cara menghilangkan semua
imbuhan (affix) yang terdiri dari awalan (prefix), sisipan (infix), akhiran (suffix) dan kombinasi awalan dan akhiran (confix).
Algoritma yang digunakan algoritma Nazief dan Andriani.
"""

factory = StemmerFactory() #Variabel factory berisi class untuk proses stemming dari library stemming
stemmer = factory.create_stemmer() #Variabel stemmer digunakan untuk mengambil fungsi pembuatan stemming dari class StemmerFactory() 

#fungsi untuk proses stemming
def stemmed_wrapper(term):
  return stemmer.stem(term)

#variabel untuk menampung dictionary dari kata-kata term stemming
term_dict = {}

"""
Proses iterasi untuk pembuatan dict termnya, yang dimana akan dilakukan pengecekan kata kata hasil dari proses tokenisasi, 
Jadi setelah ditokenisasi, kemudian semua kata-kata tersebut akan ditampung di dalam term_dict yang dimana nantinya kata-kata 
tersebut akan melalui proses dari fungsi stemming. Jika kata tersebut bukan termasuk di variabel term_dict maka list kata di 
dictionary tersebut akan dikosongkan.
"""
for document in data['sentiment_text']:
  for term in document:
    if term not in term_dict:
      term_dict[term] = " "

print(len(term_dict))
print("------------------------------")

"""
proses iterasi untuk pengecekan setiap kata di dalam variabel term_dict, yang kemudian di dalam proses tersebut terjadinya
proses perubahan setiap kata yang akan dilakukan oleh fungsi stemmed_wrapper yang berisi proses stemming kata. Setelah itu,
akan ditampilkan hasil dari proses perubahan.
"""
for term in term_dict:
  term_dict[term] = stemmed_wrapper(term)
  print(term,":",term_dict[term])

#print(term_dict)
print("-----------------------------")

"""
Kemudian hasil perubahan stemmingnya dimasukkan ke dalam dataframe untuk diolah ke tahapan berikutnya
"""
def get_stemmed_term(document):
  return[term_dict[term] for term in document]

data['sentiment_text'] = data['sentiment_text'].apply(get_stemmed_term)
print(data['sentiment_text'])
```

```python
"""
Pada tahapan ini digunakan untuk proses menghilangkan kata-kata yang tidak mendeskripsikan pada studi kasus yang semestinya 
dihilangkan. Kata yang dihilangkan adalah kata yang ada dalam dictionary yang berisi daftar kata (stopword list). 
"""

# ----------------------- ambil stopword dari library NLTK -------------------------------
#Menggunakan file package Stopword berbahasa Indonesia yang sudah di install sebelumnya dalam library NLTK
list_stopwords = stopwords.words('indonesian')

# ---------------------------- menambahkan stopword secara manual  ------------------------------------
"""
Pada bagian ini digunakan untuk menambahkan beberapa kata Stopword secara manual jika dirasa masih 
ada beberapa kata yang ingin dihapus karena tidak relevan dengan studi kasus, dan kata tersebut tidak 
terdaftar pada file stopword bawaan dari library NLTK
"""
list_stopwords.extend(['yah','toh','lah','ud','lw','wkwk',
                       'jleb','ne','wkw','xyz','rt','dong','haha',
                       'oalah','ish','hih','ngnt','ew','ups','jilidiii',
                       'hm','cuk','hadeh','au','ah','via','ae','hem','mah',
                       'wkwkwk','yg', 'dg', 'rt', 'dgn', 'ny', 'd', 'klo', 
                       'kalo', 'amp', 'biar', 'krn', 'nya', 'nih', 'sih', 'tidak',
                       'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 
                       'jd', 'jgn', 'sdh', 'aja', 'n', 't', 
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh',
                       '&amp', 'os','hadeuh','neg','cukz','seh','euy','so',
                       'cmiw','pls','bhahaha','bhaha','blak','blakan','jreng',
                       'caur','penya','b','blablabla','seh','ni','owalah','dhobol','lho',
                       'adadeh'])

# ---------------------------------------------------------------------------------------

"""
Ubah data list menjadi bertipe data set agar tidak ada elemen  dengan nilai yang sama di dalam 1 file olahan stopwords
"""
list_stopwords = set(list_stopwords)

#Fungsi untuk hapus kata kata pada daftar list yang terdaftar dalam file package Stopwords sebelumnya
def stopwords_removal(words):
    return [word for word in words if not word in list_stopwords]

data['sentiment_text'] = data['sentiment_text'].apply(stopwords_removal) 

print(data['sentiment_text'].head(20))
```

```python
#Fungsi yang digunakan untuk menggabungkan setiap kata yang terpisah akibat proses tokenisasi sebelumnya
def join_text(sentiment_text):
    return ' '.join(sentiment_text)

data['sentiment_text'] = data['sentiment_text'].apply(join_text)
print(data['sentiment_text'].head(20))
```
